### Rules

1. A variable name must start with a letter or _ and we don't have space instead we have _

```python
myvar = "AI"
my_var = "Hi"
_my_var = "python"
myVar = "in"
MYVAR = "awesome"
myvar2 = "programming"
```

2: A variable name cannot start with a number

```python
2myvar = 'hello'
```

3: Only contain alpha-numeric characters and underscores (A-z, 0-9, and _ ) and don't use symbols <>&%#@!|/

```python
&mynum = 10
```

4. Don't use keywords in python

```python
str = 'Ali'
int = 10
```
