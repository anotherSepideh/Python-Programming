## Continue and Break

#### Example3

```python
i = 0 
while i < 5:
    i += 1
    if i == 3:
        print('Continue!')
        continue
    print(i)
```

#### Example4: Game

```python
secret_word = 'python'
counter = 0
while True:
    word = input('Enter the secret word: ').lower()
    counter += 1
    if secret_word == word:
        print('Bravo!')
        break
    elif secret_word != word and counter <=7:
        print('Try Again')
    else:
        print('You lose :( ')
        break
    

```

```python
n = 0
temp = 0
result = 0
lst1 = []
while n < 5:
    temp = temp * 10 + 2
    lst1.append(temp)
    result += temp
    n += 1
print(lst1)
print(result)
```

#### Exercise3

```python
# Reverse a given integer number
# Given: 76542
# output: 24567
```

```python
num = int(input('Enter a number: '))
temp = num
reverse_num = 0
while num > 0:
    reminder = num % 10
    reverse_num = reverse_num * 10 + reminder
    num = num // 10
print(f'The reverse of {temp} is {reverse_num}')
```
